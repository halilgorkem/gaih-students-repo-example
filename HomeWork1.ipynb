```python
myList1 = [1, 3, 5, 7, 9]
myList2 = [2, 4, 6, 8, 10]
myList3 = []
myList4 = []
```

```python
myList3 = myList1 + myList2
```

```python
myList3
```

```python
for i in myList3:
    myList4.append(i*2)
```

```python
myList4.sort()
myList4
```

```python
for j in myList4:
    print("{} data type of element = {}".format(j, type(j)))
```
